TASK 13: STACKING AND CONCATENATING

```python
num1= np.array([4,5,6])
num2=np.array([7,8,9])
vstack= np.vstack(num1)
display(vstack)
vstack1=np.vstack(num2)
display(vstack1)
hstack=np.hstack(num1)
display(hstack)
hstack1=np.hstack(num2)
display(hstack1)
np.concatenate((hstack,hstack1),axis=0)
```
