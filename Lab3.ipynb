## PYTHON REQUEST LIBRARY

```python
#  using the python request library 
import json
url="http://universities.hipolabs.com/search?country=India"
response=requests.get(url) # Use the request library to get the contents from the url
Data=[]

for i in range(0,433): ## Iterate through the JSON for selected keys
    json_data= json.loads(response.text)
    name=json_data[i]['name']
    State= json_data[i]['state-province']
    Data.append([name,State])

       
```

```python
# Inserting data into a dataframe
df=pd.DataFrame(Data,columns=['Name','State'])
# Converting dataframe to csv file
df.to_csv('Data.csv', index=False)
```

## PYTHON AND BEATIFUL SOUP

```python
# Scraping the contents of a webpage using request and beautifulsoup library
books=[]
# Iterating through contents of webapge, for multiple pages
for i in range(1,3):
    url= f"https://books.toscrape.com/catalogue/category/books/mystery_3/page-{i}.html"
    response= requests.get(url)
    response= response.content
    # Parsing the contents of the webpages
    soup= BeautifulSoup(response,"html.parser")
    ol= soup.find('ol')
    articles= ol.find_all('article',class_='product_pod')
    
    # Selecting specific contents of webpage
    for article in articles:
        image=article.find('img')
        title=image.attrs['alt']
        rating= article.find('p')
        rating= rating['class'][1]
        price=article.find('p',class_='price_color').text
        price=float(price[1:])
        books.append([title,price,rating])
    
```

```python
# Inserting data into a dataframe 
df= pd.DataFrame(books,columns=['Title','Price','Rating'])
# Coverting dataframe to a csv file
df.to_csv('books.csv',index=False)
```
